```python
def Model_AR1(data):
    alpha1 = sum([data[t]*data[t-1] for t in range(1,len(data))])/sum([data[t-1]**2 for t in range(1,len(data))])
    return alpha1
```

```python
def initial_Mp(p,data):
    Mp= numpy.empty([p, p]) # 3-lignes , 2-cols
    Bp= numpy.zeros(p)
    for i in range(0,p):
        #Mp[i,i]=sum([data[t-i]**2 for t in range(p,len(data))])
        Bp[i]=  sum([data[t]*data[t-i-1] for t in range(p,len(data))])
    for i in range(0,p):
        for j in range(0,p):
            Mp[i,j]= sum([data[t-i-1]*data[t-j-1] for t in range(p,len(data))])
   
    return numpy.linalg.solve(Mp, Bp) 
    #return Mp
    
    
```
